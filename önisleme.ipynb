```python
import pandas as pd 
import os
import ast
```

```python
# "dataset_combined" klasörüne git
folder_path = "dataset_combined"
os.chdir(folder_path)
dfs = {}
# tüm CSV dosyalarını al ve oku
for filename in os.listdir():
    if filename.endswith(".csv"):
        # dosya adından DataFrame adını ve dosya adını çıkart
        df_name = filename.split(".")[0].replace(" ", "") 
        df = pd.read_csv(filename)
        exec(df_name + " = pd.DataFrame(df)") # DataFrame'i kaydet
        dfs[df_name] = df
```

```python
def str_to_dict(ilce):
    """
    bu fonksiyon çektiğimiz verilerin string olarak dönen 'Concentration' ve 'AQI' değerlerini sözlük yapısına çevirir. None değerleri 0 a eşitler 
    ek olarak hiç değer olmayan zaman dilimlerinin de oranını ve sayısını verir.
    """
    hata_sayisi = 0
    aqı_hata_sayisi = 0
    for i in range(len(ilce)):
        try:
            ilce['Concentration'][i] = ast.literal_eval(ilce['Concentration'][i].replace('None', '0'))
        except AttributeError :
            hata_sayisi += 1

        try:
            ilce['AQI'][i] = ast.literal_eval(ilce['AQI'][i].replace('None', '0'))
        except AttributeError :
            aqı_hata_sayisi += 1
        
    print(f"Hatalı dönüşüm: {hata_sayisi}")
    oran = hata_sayisi/len(ilce) *100
    print(f"oranı: {oran}")
    
    print(f"Hatalı dönüşüm: {aqı_hata_sayisi}")
    oran = aqı_hata_sayisi/len(ilce) *100
    print(f"oranı: {oran}")
    
```

```python
def dict_to_columns(df):
    concentration_df = df['Concentration'].apply(pd.Series)
    concentration_df.columns = concentration_df.columns.str.upper()
    df = pd.concat([df, concentration_df], axis=1)
    df.drop('Concentration', axis=1, inplace=True)

    AQI_df = df['AQI'].apply(pd.Series)
    AQI_df.columns = AQI_df.columns.str.upper()+"_AQI"
    df = pd.concat([df, AQI_df], axis=1)
    df.drop('AQI', axis=1, inplace=True)

    return df
```

```python
#Exec ile dosyaları tekrar kaydetmek mümkün müdür 
for df_name, df in dfs.items():
    print(df_name+" sözlük yapısı oluşturuluyor")
    str_to_dict(df)
    print(df_name+" df güncelleniyor")
    df = dict_to_columns(df)
    exec(f"{df_name} = df")
    print(df_name+" tamamlandı")
```
